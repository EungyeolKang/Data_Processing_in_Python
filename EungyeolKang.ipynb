```python
#Part1_Q2
import pandas as pd

akam = pd.read_csv("AKAM.csv")
keys = pd.read_csv("KEYS.csv")
wab = pd.read_csv("WAB.csv")
wynn = pd.read_csv("WYNN.csv")

def highest_price(data):
    highest_pr = data['Close'].max()
    print(highest_pr)

def lowest_price(data):
    lowest_pr = data['Close'].min()
    print(lowest_pr)

highest_price(akam)
lowest_price(akam)

highest_price(keys)
lowest_price(keys)

highest_price(wab)
lowest_price(wab)

highest_price(wynn)
lowest_price(wynn)
```

```python
#Part1_Q3

import numpy as np

def log_return(data):
    data['log_ret'] = np.log(data['Close']/data['Close'].shift(1))

    logmin = data['log_ret'].min()
    logmax = data['log_ret'].max()
    logmean = data['log_ret'].mean()
    logmedian = data['log_ret'].median()

    print(logmin, logmax, logmean, logmedian)

log_return(akam)
log_return(keys)
log_return(wab)
log_return(wynn)
```

```python
#Part1_Q4

def highest_gain(data):
    highest_g = data.loc[data['log_ret'].idxmax()], ['Date', 'log_ret']
    print(highest_g)

def highest_loss(data):
    highest_l = data.loc[data['log_ret'].idxmin()], ['Date', 'log_ret']
    print(highest_l)

highest_gain(akam)
highest_loss(akam)

highest_gain(keys)
highest_loss(keys)

highest_gain(wab)
highest_loss(wab)

highest_gain(wynn)
highest_loss(wynn)
```

```python
#Part1_Q5

def average_volume(data):
    data['Date'] = pd.to_datetime(data['Date'])
    
    data.set_index('Date', inplace=True)
    
    weekly_volume = data.groupby('Company').resample('W')['Volume'].mean()
    
    weekly_volume = weekly_volume.reset_index()

    weekly_volume_res = average_volume(data)
    print(weekly_volume_res)
```
